```python
import pandas as pd
# Read in County-Level csv and convert to ready-to-use dataframe.
path_to_county_data = './County_Level_Data.csv'
county_df = pd.read_csv(path_to_county_data)
# Read in State-Level csv and convert to ready-to-use dataframe.
path_to_state_data = './State_Level_Data.csv'
state_df = pd.read_csv(path_to_state_data)
# Read in County health indicators
path_to_all_counties = './county_health_rankings_data.csv'
all_df = pd.read_csv(path_to_all_counties)
```

# Load Dataframes into SQLITE Database

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///HealthIndicatorsData.db')
county_df.to_sql('county_data', con=engine, if_exists='replace',index_label="id")
state_df.to_sql('state_data', con=engine, if_exists='replace',index_label="id")
all_df.to_sql('county_health_rankings_data', con=engine, if_exists='replace',index_label="id")
```

```python
engine.execute("SELECT * FROM county_data WHERE State='North Carolina' LIMIT 3").fetchall()
```

```python
engine.execute("SELECT * FROM state_data WHERE State='Tennessee' LIMIT 3").fetchall()
```
